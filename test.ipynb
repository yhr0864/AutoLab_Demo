# 基类

```python
from dataclasses import dataclass, asdict
from typing import Optional

@dataclass
class Device:
    device_id: str
    name: str
    model: str
    manufacturer: Optional[str] = None
    location: Optional[str] = None
```

# 子类

```python
@dataclass
class Centrifuge(Device):
    max_speed_rpm: int = 0
    max_capacity_ml: float = 0.0
    refrigerated: bool = False

    def __post_init__(self):
        if self.max_speed_rpm <= 0:
            raise ValueError("最大转速必须大于 0")
        if self.max_capacity_ml <= 0:
            raise ValueError("最大容量必须大于 0")

@dataclass
class Incubator(Device):
    min_temp: float = 0.0
    max_temp: float = 0.0
    volume_l: float = 0.0

    def __post_init__(self):
        if self.min_temp >= self.max_temp:
            raise ValueError("最小温度必须小于最大温度")
        if self.volume_l <= 0:
            raise ValueError("容积必须大于 0")
```

# 实例

```python
c = Centrifuge(
    device_id="C001",
    name="冷冻离心机",
    model="CF-8000",
    manufacturer="ABC",
    location="3楼样品室",
    max_speed_rpm=18000,
    max_capacity_ml=200,
    refrigerated=True
)

i = Incubator(
    device_id="I001",
    name="恒温培养箱",
    model="INC-300",
    manufacturer="XYZ",
    location="2楼微生物室",
    min_temp=5,
    max_temp=70,
    volume_l=150
)
```
