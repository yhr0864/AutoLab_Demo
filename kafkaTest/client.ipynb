# 查看所有Topic及每个Topic的详细消息

```python
from kafka.admin import KafkaAdminClient

# 创建 KafkaAdminClient 实例
admin_client = KafkaAdminClient(bootstrap_servers='localhost:9092', client_id='admin-client')

# 获取所有 topic 列表
topics = admin_client.list_topics()
print("所有 Topic:", topics)
results = admin_client.describe_topics(topics)
for detail in results:
    if detail['topic'] == "__consumer_offsets":
        continue
    print("Topic 详情:", detail)

admin_client.close()
```

# 创建消费者

```python
from kafka import KafkaConsumer

consumer = KafkaConsumer('school', bootstrap_servers='localhost:9092')
for message in consumer:
    print(f'Received message: {message.value.decode("utf-8")}')
```

# 创建生产者

```python
from kafka import KafkaProducer

producer = KafkaProducer(bootstrap_servers='localhost:9092')
for i in range(10):
    message = f'Hello {i} from Kafka.'.encode('utf-8')
    producer.send(topic='school', value=message)
producer.close()
```
